```python
import msprime
import matplotlib.pyplot as plt
```

```python
# set sample size and number of trials
sps=2
num=10000
```

```python
# define the function that finds the least coalescent time
def t_mrca(foo):
    # reverse table
    bar = foo[::-1]
    # number of symbols per line
    baz = len(bar)//bar.count("\n")
    # index of the first blank,
    # the blank's length may vary,
    # and a single tree doesn't have a blank
    tiz = len(bar)
    for idx in range(4*sps+1,6*sps+1):
        if bar.count("┊"+idx*" "+"┊")!=0:
            if bar.index("┊"+idx*" "+"┊")<tiz:
                tiz = bar.index("┊"+idx*" "+"┊")
    # the time before the blank 
    tud = tiz//baz-1
    tus = bar.index(" 0 ")+2
    qux = bar[baz*tud+tus:baz*(tud+1)]
    wib = float(qux[::-1])
    return(wib)
```

```python
# run the simulation num times
t_mrca_list = []
for idxx in range(num):
    hoge = msprime.sim_ancestry(
        samples=sps,
        recombination_rate=1e-8,
        sequence_length=5000,
        population_size=10000,
        record_full_arg=True)
    piyo = hoge.draw_text()
    t_mrca_list.append(t_mrca(piyo))
```

```python
# plot the distribution
plt.hist(t_mrca_list, density=False, bins=num//400)
plt.ylabel("Number of Coalesences")
plt.xlabel("Fastest Coalescent Time (?)")
```
